```python
import matplotlib.pyplot as plt
import numpy as np
import csv
```

```python
# Description:   Function to use and iterate the logistic map equation
# @param rate    Growth rate used by the logistic map equation
# @param x       The current year population value used by the logistic map equation
# @return:       Returns the next iteration of the logistic map equation
# assume x between 0 and 1
def logistic_eqn(rate,x):
    return rate * x * (1 - x)
```

```python
# Description:   Function to use and iterate the logistic map equation with power on the restricting
#                term of the logistic map equation
def logistic_eqn_power(rate,x,n):
    return rate * x * ((1 - x)**n)
```

```python
# Creates 50 (default) evenly spaced numbers between interval of 0 to 1
x =  np.linspace(0,1)
# Creates y values for logistic eqn with rate of 2 and x values between 0 and 1
y = logistic_eqn(2,x)
# Plots the current value vs next iteration with a growth rate of 2
fig, ax = plt.subplots()
ax.plot(x,y)
ax.set_xlabel('xo')
ax.set_ylabel('xn')
ax.set_title('r = 2')
plt.show()
```

```python
# Creates y values for logistic eqn with rate of 3 and x values between 0 and 1
y = logistic_eqn(3,x)
# Plots the current value vs next iteration with a growth rate of 3
fig, ax = plt.subplots()
ax.plot(x,y)
ax.set_xlabel('xo')
ax.set_ylabel('xn')
ax.set_title('r=3')
plt.show()
```

```python
# Sets up paramters for recreating bifurcation diagram
# Creates an array n rate values between
n = 1000
r = np.linspace(0,4,n)
iterations = 500
x = 0.5

fig, ax = plt.subplots()
for i in range(iterations):
    x = logistic_eqn(r,x)
    ax.plot(r,x)
ax.set_title('Bifurcation Diagram')
ax.set_xlabel('Rate')
ax.set_ylabel('Equilibrium Population')
plt.show()
```

```python
# Creates a plot of logistic_eqn_power for 100 n values from 0.1 to 10
# in order to determine the effects of changing this term
fig, ax = plt.subplots()
x = np.linspace(0,1)
for i in range(1,100):
    y = logistic_eqn_power(2,x,i/10)
    ax.plot(x,y)
ax.set_title('(1-x)^n')
ax.set_xlabel('xo')
ax.set_ylabel('xn')
plt.show()
```

```python
# Initialize arrays for years, hares and lynx
years = []
hares = []
lynx = []

# Read in data from HudsonBayData and put data in their respective arrays
with open('HudsonBayData.txt') as HusonBayDataFile:
    csv_reader = csv.reader(HusonBayDataFile, delimiter=',')
    line_count = 0
    for row in csv_reader:
        if line_count == 0:
            line_count += 1
        else:
            years.append(row[0])
            hares.append(int(row[1]))
            lynx.append(int(row[2]))
            line_count += 1
            
#Create figure for Predator Prey relations
plt.figure(figsize=(15,10))
plt.plot(years,hares,label='Hares')
plt.plot(years,lynx,label='Lynx')
plt.ylim(0,150)
plt.title('Predator Prey Relations')
plt.ylabel('Population (thousands)')
plt.xlabel('Year')
plt.xticks(rotation=50)
plt.legend(loc='best')
plt.show()
```
